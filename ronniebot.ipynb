```python
#packages need to installed 
#1)pyttsx3 
#library for getting responce from audio.
#search by typing pptsx3 and entire documentation you will get

#2)pyaudio(if you are using microphone for taling audio as input)
#library to ask question in the form of audio 
#installation done by 
#pip install pipwi
#pipwin install pyaudio


#Import library for bot with data
#Library: chatterbot
#Class  : ChatBot
from chatterbot import ChatBot

#Import the algorithm (trainer)
#2. UbuntuCorpusTrainer
#Library: chatterbot
#Module : trainers
#Class  : ChatterBotCorpusTrainer
from chatterbot.trainers import ChatterBotCorpusTrainer


#define bot name and train the bot 
bot=ChatBot('ronnie')
bot.set_trainer(ChatterBotCorpusTrainer)
bot.train('chatterbot.corpus.english')


#importing the module -tkinter(for python 3 and above)
#for python 2 and above it is Tkinter

import tkinter          #tkinter is installed by default when you download python all we need to do is just import the library

from tkinter import *   # * means import everything 

import pyttsx3 as pt    # this is a library which converts text data as input to audio form as output
                        # reference:- pypi.org .To install this library you need to run the command :-pip install pyttsx3 


from tkinter import messagebox
import speech_recognition as sp #Library to perform speech recognition as input 
                                #install using command pip install speechRecognition
import threading


#we first create a blank main window with options like close,maximize,minimize buttons on the top of the page

window=tkinter.Tk()
#give the size of the window using geometry(parameters)
window.geometry("600x700")

#title for the window
window.title('AUTOBOT')

#add image to your window
pic=PhotoImage(file='bot.png')
photo=Label(window,image=pic)
photo.pack(pady=3)

#creation of frame
#frames allow to organize and group widgets 
frame=Frame(window, width = 300 ,height=100) #assing frame class to parent window

#creates a scrollbar(widget) to the frame
sc=Scrollbar(frame)



#add text to your window like a heading using Label(parameters)
#lb variable
lb=Label(frame,text='>|< AUTOBOT >|<',font=("Arial Bold",20))
lb.pack()
#instead of pack() we can use grid to align the text at particular position 
# eg :- lb.grid(column=0,row=0_)


#creates a listbox 
message=Listbox(frame,width=120,height=12,yscrollcommand=sc.set,bg='Yellow')


#fill the widget(scrollbar along y axis only to the rightside)
sc.pack(side=RIGHT,fill=Y)
message.pack(side=LEFT,fill=BOTH,pady=10)
frame.pack()

#text field entry
#entry(parameter) allows us to enter text (ask questions)
text=Entry(window,font=("Arial",20),bd=5)
text.pack(fill=X,pady=30)

#steps to get reply from bot in the form of audio
#reference from pyttsx3 documentation
engine=pt.init()
voices=engine.getProperty('voices')
print(voices)
engine.setProperty('voice',voices[1].id)#voice[1]=female voice, voice[0]=male voice

#Talk function is made to get responce from bot in the form of audio
def talk(word):
    engine.say(word)
    engine.runAndWait()

#function to get response from bot as both text and audio by calling talk()function
def hello():
        data=text.get()
        answer=bot.get_response(data)
        message.insert(END,"You  :", data)
        message.insert(END,"AUTOBOT :",str(answer))
        talk(answer)
        text.delete(0,END)
        message.yview(END)#automatically scrolls down 


#this button allows us to  get respond from a question asked from the bot
b=Button(window,text="ASK",font=("Arial",20),command=hello)
b.pack()


#this function allows us to get reply from bot by pressing enter key button   
def sample(event):
    b.invoke()
#joins main window with enter key button    
window.bind('<Return>',sample)



#function which takes our audio as input with different functions like Microphone(),Recognizer()
#converts the audio to string data type
def takeaudio():
    speech=sp.Recognizer()          #recognize our audio
    speech.pause_threshold=1         
    print("I am listening")     #to indicate that the bot is hearing 
    with sp.Microphone() as m:
        try:
            audio=speech.listen(m)
            st=speech.recognize_google(audio,language='eng-in')#Convert audio to string
            print(st)
            text.delete(0,END)
            text.insert(0,st)      #insert data to text field
            hello()
        except Exception as e:
            print(e)
            print("cant respond")

#repeat the process of taking input from audio n number of times
def repeat():
    while True:
        takeaudio()

        

#create messagebox showing warning message when you press close button(x)
canvas = tkinter.Canvas(window, width = 300, height = 300)
canvas.pack()

#creating a function Exitwindow()
def Exitwindow():
    Box = tkinter.messagebox.askquestion ('EXIT','Are you sure you want to exit the application',icon = 'warning')
    if Box == 'yes':
       window.destroy()
    else:
        tkinter.messagebox.showinfo('Return',' Returning back')


#this button represents exit button which asks us that do you want to quit yes or no        
button = tkinter.Button (window, text='EXIT',command=Exitwindow,bg='white',fg='red',width=20,height=2)
canvas.create_window(150, 50, window=button)


#GUI threads(tasks,function calls) will be busy with takeaudio() function and the GUI interface won't be visible
#so we add takeaudio() to another thread 
t=threading.Thread(target=repeat)
t.start()

#allows the main window to run continously using mainloop() function
window.mainloop()
```
